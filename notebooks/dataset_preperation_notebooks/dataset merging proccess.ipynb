```python
import pandas as pd
import seaborn as sns
```

```python
index_col = "open_time"
kline_columns = ["open", "high", "low", "close", "volume", "close_time", "quote_asset_volume", "number_of_trades", "taker_buy_base_asset_volume","taker_buy_quote_asset_volume", "ignore"]
```

```python
kline_df = pd.read_csv("~/diploma_data/BTCUSDT-1s-2021-04.csv", names=kline_columns)
kline_df.index.name = index_col
```

```python
kline_df = kline_df[kline_df.columns[kline_df.columns.isin(["open", "high", "low", "close", "volume"])]]
```

```python
lob_df = pd.read_csv("~/diploma_data/BTC_1sec.csv", index_col=0)
```

```python
lob_df["system_time"] = (pd.to_datetime(lob_df["system_time"]).dt.tz_localize(None) - pd.Timestamp("1970-01-01")) // pd.Timedelta('1s')*1000
```

```python
lob_df = lob_df.set_index("system_time")
```

```python
lob_df.head()
```

```python
lob_df.columns
```

# SKIP

```python
lob_columns = lob_df.columns[lob_df.columns.str.contains('midpoint|spread|buys|sells|(bids|asks)_(distance|limit_notional)_\d+')]
lob_columns
```

```python
lob_df = lob_df[lob_columns]
lob_df.info()
```

```python
new_columns = []

for col in lob_df.columns:
    if 'distance' in col:
        new_col_name = col.replace('distance', 'level')
#         lob_df[col] = lob_df.midpoint + lob_df[col]
    elif 'limit_notional' in col:
        new_col_name = col.replace('limit_notional', 'volume')
    else:
        new_col_name = col
        
    new_columns.append(new_col_name)

lob_df.columns = new_columns
```

# SKIP END

```python
merged = pd.merge(kline_df, lob_df, left_index=True, right_index=True)
merged.info()
```

```python
# calculate the correlation matrix
corr = merged.corr()

# plot the heatmap
sns.heatmap(corr, 
        xticklabels=corr.columns,
        yticklabels=corr.columns)
```
